```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python

def read_df_from_csv(path: str) -> pd.DataFrame:
    return pd.read_csv(
        filepath_or_buffer=path,
        sep=',',
        header=0,
        dtype={
        'ORIGIN_AIRPORT': str,
        'DESTINATION_AIRPORT': str,
        'SCHEDULED_DEPARTURE': str,
        'DEPARTURE_TIME': str,
        'SCHEDULED_ARRIVAL': str,
        'ARRIVAL_TIME': str
    },
    low_memory=True
    )
```

```python
def fix_hhmm(hhmm_series, year, month, day):
    hhmm_series = hhmm_series.fillna(0).astype(int)

    overflow = hhmm_series == 2400
    hhmm_series[overflow] = 0

    hh = hhmm_series // 100
    mm = hhmm_series % 100

    dt = pd.to_datetime(dict(year=year, month=month, day=day)) + pd.to_timedelta(hh, unit='h') + pd.to_timedelta(mm, unit='m')

    dt[overflow] += pd.Timedelta(days=1)

    return dt

flights_df['SCHEDULED_DEPARTURE_dt'] = fix_hhmm(
    flights_df['SCHEDULED_DEPARTURE'],
    flights_df['YEAR'].astype(int),
    flights_df['MONTH'].astype(int),
    flights_df['DAY'].astype(int)
)
```

```python

time_cols = ['SCHEDULED_DEPARTURE','DEPARTURE_TIME','SCHEDULED_ARRIVAL','ARRIVAL_TIME', 'WHEELS_OFF', 'WHEELS_ON']

for col in time_cols:
    valid_mask = flights_df[col].notna()

    hhmm = flights_df.loc[valid_mask, col].astype(int).astype(str).str.zfill(4)
    hours = hhmm.str[:2].astype(int)
    minutes = hhmm.str[2:].astype(int)

    flights_df.loc[valid_mask, col + '_dt'] = (
        pd.to_datetime(
            pd.DataFrame({
                'year': flights_df.loc[valid_mask, 'YEAR'],
                'month': flights_df.loc[valid_mask, 'MONTH'],
                'day': flights_df.loc[valid_mask, 'DAY']
            })
        ) + pd.to_timedelta(hours, unit='h') + pd.to_timedelta(minutes, unit='m')
    )

for col in ['SCHEDULED_DEPARTURE_dt', 'DEPARTURE_TIME_dt', 'SCHEDULED_ARRIVAL_dt', 'ARRIVAL_TIME_dt', 'WHEELS_OFF_dt', 'WHEELS_ON_dt']:
    flights_df[col + '_min'] = flights_df[col].dt.hour * 60 + flights_df[col].dt.minute

flights_df.drop(columns=time_cols)
```

```python
## Data Cleaning / Missing Value Imputation

cancellation_map = {
    'A': 'Airline/Carrier',
    'B': 'Weather',
    'C': 'National Air System',
    'D': 'Security'
}

delays_cols = [
    "AIR_SYSTEM_DELAY",
    "SECURITY_DELAY",
    "AIRLINE_DELAY",
    "LATE_AIRCRAFT_DELAY",
    "WEATHER_DELAY"
]

flights_df['TAIL_NUMBER'] = flights_df['TAIL_NUMBER'].fillna(value='UNKNOWN')
flights_df['CANCELLATION_REASON'] = flights_df['CANCELLATION_REASON'].map(cancellation_map).fillna('Not cancelled')

for delay_col in delays_cols:
    flights_df[delay_col] = flights_df[delay_col].fillna(value=0)
```
